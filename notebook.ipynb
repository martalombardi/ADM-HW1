```python
import os

def print_tree(root=".", max_depth=2):
    root = os.path.abspath(root)
    for dirpath, dirnames, filenames in os.walk(root):
        depth = dirpath.replace(root, "").count(os.sep)
        if depth > max_depth:
            continue
        indent = "  " * depth
        print(f"{indent}{os.path.basename(dirpath)}/")
        for f in filenames:
            print(f"{indent}  {f}")

print_tree(".", max_depth=3)
```

```python
# ============================================================
# 0) Setup utility
# ============================================================
from pathlib import Path
import os
import re
import subprocess
from collections import Counter

REPO_ROOT = Path(".")   # esegui dalla root del repo clonato

def show_file(path, n_lines=200, start_line=1):
    path = Path(path)
    print(f"\n===== {path} =====\n")
    if not path.exists():
        print(f"[ERRORE] File non trovato: {path}")
        return
    with open(path, "r", errors="ignore") as f:
        for i, line in enumerate(f, start=1):
            if i < start_line:
                continue
            if i >= start_line + n_lines:
                print("\n... (troncato)")
                break
            print(f"{i:4d}: {line.rstrip()}")

def grep(pattern, root=".", glob="*"):
    root = Path(root)
    matches = []
    regex = re.compile(pattern)
    for p in root.rglob(glob):
        if not p.is_file():
            continue
        try:
            with open(p, "r", errors="ignore") as f:
                for i, line in enumerate(f, start=1):
                    if regex.search(line):
                        matches.append((str(p), i, line.rstrip()))
        except Exception:
            pass
    return matches

def print_matches(matches, max_rows=50):
    for j, (path, line_no, line) in enumerate(matches[:max_rows], start=1):
        print(f"{j:3d}. {path}:{line_no}: {line}")
    if len(matches) > max_rows:
        print(f"... altri {len(matches)-max_rows} match")

def list_cpp_h_files(root="."):
    root = Path(root)
    files = sorted([str(p) for p in root.rglob("*") if p.suffix in {".cpp", ".h"}])
    for f in files:
        print(f)
    print(f"\nTotale file .cpp/.h: {len(files)}")
```

```python
# ============================================================
# 1) Elenco file sorgente reali
# ============================================================
src_files = []
for folder in ["TreeRePair", "TreeRePairLibrary", "TreeRePairCommonsLibrary", "TreeDePair", "TreeDePairLibrary"]:
    p = REPO_ROOT / folder
    if p.exists():
        for f in p.rglob("*"):
            if f.suffix in {".cpp", ".h"}:
                src_files.append(str(f))

for f in sorted(src_files):
    print(f)

print("\nTotale:", len(src_files))
```

```python
# ============================================================
# 3) Dove viene definito il core?
# ============================================================
patterns = [
    r"class TreeRePairAlgorithm",
    r"TreeRePairAlgorithm::TreeRePairAlgorithm",
    r"TreeRePairAlgorithm::~TreeRePairAlgorithm",
    r"void TreeRePairAlgorithm::parseInput",
    r"void TreeRePairAlgorithm::pruneGrammar",
    r"void TreeRePairAlgorithm::pruneDagGrammar",
]

for pat in patterns:
    print(f"\n===== PATTERN: {pat} =====")
    matches = grep(pat, root="TreeRePairLibrary")
    print_matches(matches, max_rows=20)
```

```python
# ============================================================
# 5) Include graph minimale
# ============================================================
def extract_includes(path):
    path = Path(path)
    includes = []
    with open(path, "r", errors="ignore") as f:
        for line in f:
            line = line.strip()
            if line.startswith("#include"):
                includes.append(line)
    return includes

for fp in [
    "TreeRePair/TreeRePair.cpp",
    "TreeRePairLibrary/TreeRePairAlgorithm.cpp",
    "TreeRePairLibrary/DigramReplacer.cpp",
    "TreeRePairLibrary/DigramCounter.cpp",
    "TreeRePairLibrary/OccurrenceFinder.cpp",
]:
    if Path(fp).exists():
        print(f"\n===== INCLUDES: {fp} =====")
        for inc in extract_includes(fp):
            print(inc)
```

```python
# ============================================================
# 6) Cerca strutture dati critiche per la RAM
# ============================================================
critical_patterns = [
    r"HASH_D_D",
    r"DigramPriorityQueue",
    r"DigramCounter\s*\*",
    r"vector<\s*TreeNode\s*\*>",
    r"unordered_map",
    r"new\s+TreeNode",
    r"new\s+Digram",
    r"occurrences?",
    r"non-overlapping occurrences",
]

for pat in critical_patterns:
    print(f"\n===== SEARCH: {pat} =====")
    matches = grep(pat, root="TreeRePairLibrary")
    matches += grep(pat, root="TreeRePairCommonsLibrary")
    print_matches(matches, max_rows=40)
```

```python
# ============================================================
# 7) Ranking dei file più "centrali" per TreeRePair
# ============================================================
terms = [
    "digram", "occurrence", "replace", "priority", "grammar",
    "TreeNode", "Production", "parse", "dag", "prune"
]

scores = Counter()

for f in src_files:
    try:
        text = Path(f).read_text(errors="ignore").lower()
        for t in terms:
            scores[f] += text.count(t.lower())
    except Exception:
        pass

for f, s in scores.most_common(20):
    print(f"{s:4d}  {f}")
```

```python
# ============================================================
# 8) Estrai firme funzioni/metodi dai file chiave
# ============================================================
def extract_signatures(path):
    path = Path(path)
    sigs = []
    pattern = re.compile(r'^\s*([A-Za-z_][\w:<>\s*&]+)\s+([A-Za-z_][\w:]*)\s*\(([^;{}]*)\)\s*(const)?\s*[{;]')
    with open(path, "r", errors="ignore") as f:
        for i, line in enumerate(f, start=1):
            if "::" in line or "(" in line:
                if pattern.search(line):
                    sigs.append((i, line.rstrip()))
    return sigs

for fp in [
    "TreeRePairLibrary/TreeRePairAlgorithm.cpp",
    "TreeRePairLibrary/DigramReplacer.cpp",
    "TreeRePairLibrary/DigramCounter.cpp",
    "TreeRePairLibrary/OccurrenceFinder.cpp",
]:
    if Path(fp).exists():
        print(f"\n===== SIGNATURES: {fp} =====")
        for line_no, line in extract_signatures(fp)[:80]:
            print(f"{line_no:4d}: {line}")
```
